```python
df = df.drop(['Evaporation','Sunshine','Cloud9am','Cloud3pm','Location','Date'],axis=1)
```

```python
df.head()
```

```python
df = df.dropna(axis=0)
```

```python
df.columns
```

```python
df.shape
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['WindGustDir']=le.fit_transform(df['WindGustDir'])
df['WindDir9am']=le.fit_transform(df['WindDir9am'])
df['WindDir3pm']=le.fit_transform(df['WindDir3pm'])
df['RainToday']=le.fit_transform(df['RainToday'])
df['RainTomorrow']=le.fit_transform(df['RainTomorrow'])
```

```python
x = df.drop(['RainTomorrow'],axis=1)
y = df['RainTomorrow']
```

```python
 sns.scatterplot(x='MaxTemp',y='MinTemp',hue='RainTomorrow',palette='inferno',data=df)
```

```python
sns.scatterplot(x='Humidity9am',y='Temp9am',hue='RainTomorrow',palette='Accent', data = df)
```

```python
plt.figure(figsize=(10,5))
sns.heatmap(df.corr(),annot=True)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.metrics import accuracy_score,r2_score
```

# Linear Regression

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
model = lr.fit(x_train,y_train)
```

```python
model.score(x,y)
```

```python
y_pred = lr.predict(x_test)
```

```python
y_pred
```

```python
r2_score(y_test,y_pred).round(2)
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
```

```python
y_pred=lr.predict(x_test)
```

```python
data = pd.DataFrame({'Actual':y_test,'Predicted':y_pred})
```
